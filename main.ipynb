```python
import pandas as pd

csv_file_path = "C:/Users/santh/Documents/Guvi\Visual Studio/Retail Order Data Analysis Project/orders.csv"

df = pd.read_csv(csv_file_path, index_col=False)

df
```

```python
df.rename(columns={'Order Id': "Order_Id"}, inplace=True)
df.rename(columns={'Order Date': "Order_Date"}, inplace=True)
df.rename(columns={'Ship Mode': "Ship_Mode"}, inplace=True)
df.rename(columns={'Postal Code': "Postal_Code"}, inplace=True)
df.rename(columns={'Sub Category': "Sub_Category"}, inplace=True)
df.rename(columns={'Product Id': "Product_Id"}, inplace=True)
df.rename(columns={'cost price': "Cost_price"}, inplace=True)
df.rename(columns={'List Price': "List_Price"}, inplace=True)
df.rename(columns={'Discount Percent': "Discount_Percent"}, inplace=True)
```

```python
df.info()
```

```python
ship_mode = df["Ship_Mode"].mode()[0]

df["Ship_Mode"] = df["Ship_Mode"].fillna(ship_mode)
```

```python
df.isnull().sum()
```

```python
df["Discount"] = (df["List_Price"] * (df["Discount_Percent"] / 100))
```

```python
df["Sales_Price"] = df["List_Price"] - (df["Discount"])
```

```python
df["Profit"] = df["Sales_Price"]-df["Cost_price"]
```

```python
import pandas as pd
import sqlalchemy as sa
from sqlalchemy import create_engine
from sqlalchemy import text

sql_con = sa.engine.URL.create(
    drivername = 'mysql+pymysql',
    username = "root",
    password = "mysql4505",
    host = "127.0.0.1",
    database ="retail_orders"
)

sql_engine = create_engine(sql_con)
```

```python
query = '''SELECT * FROM retail_orders.retail;'''

df=pd.read_sql_query(query,sql_engine)
df.head()
```

```python
#1. Find top 10 highest revenue generating products

query = text(''' SELECT 
             Product_Id, 
             Category, 
             Profit  FROM retail_orders.retail
             Order By Profit Desc limit 10;''')

df=pd.read_sql_query(query,sql_engine)
df
```

```python
#2. Find the top 5 cities with the highest profit margins

query = text(''' SELECT City, Profit 
             FROM retail_orders.retail 
             Order By Profit Desc 
             limit 5;''')

df=pd.read_sql_query(query,sql_engine)
df

```

```python
#3. Calculate the total discount given for each category

query = text(''' SELECT DISTINCT
             Category,sum(Discount) as Total_Discount 
             FROM retail 
             group by Category ;''')

df=pd.read_sql_query(query,sql_engine)
df
```

```python
#4. Find the average sale price per product category

query = text(''' SELECT DISTINCT
             Category, avg(Sales_Price) as Average_Sale_Price
             FROM retail group by Category''')

df=pd.read_sql_query(query,sql_engine)
df
```

```python
#5. Find the region with the highest average sale price

query = text(''' SELECT DISTINCT
             Region, avg(Sales_Price) as Average_Sale_Price
             FROM retail group by Region Order by Average_Sale_Price Desc;''')

df=pd.read_sql_query(query,sql_engine)
df
```

```python
#6. Find the total profit per category

query = text(''' SELECT DISTINCT
             category, sum(Profit) as Total_Profit
             FROM retail group by category;''')

df=pd.read_sql_query(query,sql_engine)
df
```

```python
#7. Identify the top 3 segments with the highest quantity of orders

query = text(''' SELECT DISTINCT
             Segment, max(Quantity) as Highest_quantity
             FROM retail 
             Group by Segment Order by Highest_quantity desc limit 3''')

df=pd.read_sql_query(query,sql_engine)
df

```

```python
#8. Determine the average discount percentage given per region

query = text(''' SELECT DISTINCT
             Region, avg(Discount_Percent) as Avg_Discount_Percentage
             FROM retail 
             Group By Region;''')

df=pd.read_sql_query(query,sql_engine)
df
```

```python
#9. Find the product category with the highest total profit

query = text(''' SELECT DISTINCT
             Sub_Category as Product, max(Profit) as Total_Profit
             FROM retail 
             Group By Product Order By Total_Profit Desc limit 1;''')

df=pd.read_sql_query(query,sql_engine)
df
```

```python
#10. Calculate the total revenue generated per year

query = text(''' SELECT DISTINCT
             Year(Order_Date) as Year , sum(Profit) as Total_Revenue
             FROM retail 
             Group By Year;''')

df=pd.read_sql_query(query,sql_engine)
df
```

```python
import pymysql
import csv

def insert_csv_to_mysql(csv_file_path):
    connection = None  # Initialize the connection variable
    try:
        # Connect to the MySQL database
        connection = pymysql.connect(
            host='127.0.0.1',          # Replace with your host
            user='root',               # Replace with your MySQL username
            password='mysql4505',        # Replace with your MySQL password
            database='retail_orders'   # Replace with your database name
        )
        cursor = connection.cursor()

        # SQL query to insert data, using INSERT IGNORE to avoid duplicate primary key errors
        sql_query = """
            INSERT INTO Product (Product_Id,Order_Id, Category, Sub_Category, Cost_price, List_Price, Quantity, Discount_Percent, Discount, Sales_Price, Profit)
            VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s,%s)
        """

        # Open the CSV file and read its contents
        with open(csv_file_path, mode='r') as file:
            csv_data = csv.reader(file)
            next(csv_data)  # Skip the header row

            # Insert each row into the table
            for row in csv_data:
                cursor.execute(sql_query, row)

        # Commit the transaction
        connection.commit()
        print(f"Data from {csv_file_path} inserted successfully into Products table")

    except Exception as e:
        print(f"Error: {e}")
    finally:
        # Close the connection only if it exists and is open
        if connection and connection.open:
            connection.close()
            print("MySQL connection closed")

# File path to the CSV file
csv_file_path = "C:/Users/santh/Documents/Guvi/Visual Studio/Retail Order Data Analysis Project/products.csv"

# Call the function to insert data
insert_csv_to_mysql(csv_file_path)


```
