```python
import pandas as pd
import sqlalchemy as sa
from sqlalchemy import create_engine
from sqlalchemy import text

sql_con = sa.engine.URL.create(
    drivername = 'mysql+pymysql',
    username = "root",
    password = "mysql",
    host = "127.0.0.1",
    database ="retail_orders"
)

sql_engine = create_engine(sql_con)
```

```python
query = '''SELECT * FROM retail_orders.retail;'''

df=pd.read_sql_query(query,sql_engine)
df.head()
```

```python
#1. Find top 10 highest revenue generating products

query = text(''' SELECT 
             Product_Id, 
             Category, 
             Profit  FROM retail_orders.retail
             Order By Profit Desc limit 10;''')

df=pd.read_sql_query(query,sql_engine)
df
```

```python
#2. Find the top 5 cities with the highest profit margins

query = text(''' SELECT City, Profit 
             FROM retail_orders.retail 
             Order By Profit Desc 
             limit 5;''')

df=pd.read_sql_query(query,sql_engine)
df

```

```python
#3. Calculate the total discount given for each category

query = text(''' SELECT DISTINCT
             Category,sum(Discount) as Total_Discount 
             FROM retail 
             group by Category ;''')

df=pd.read_sql_query(query,sql_engine)
df
```

```python
#4. Find the average sale price per product category

query = text(''' SELECT DISTINCT
             Category, avg(Sales_Price) as Average_Sale_Price
             FROM retail group by Category''')

df=pd.read_sql_query(query,sql_engine)
df
```

```python
#5. Find the region with the highest average sale price

query = text(''' SELECT DISTINCT
             Region, avg(Sales_Price) as Average_Sale_Price
             FROM retail group by Region Order by Average_Sale_Price Desc;''')

df=pd.read_sql_query(query,sql_engine)
df
```

```python
#6. Find the total profit per category

query = text(''' SELECT DISTINCT
             category, sum(Profit) as Total_Profit
             FROM retail group by category;''')

df=pd.read_sql_query(query,sql_engine)
df
```

```python
#7. Identify the top 3 segments with the highest quantity of orders

query = text(''' SELECT DISTINCT
             Segment, max(Quantity) as Highest_quantity
             FROM retail 
             Group by Segment Order by Highest_quantity desc limit 3''')

df=pd.read_sql_query(query,sql_engine)
df

```

```python
#8. Determine the average discount percentage given per region

query = text(''' SELECT DISTINCT
             Region, avg(Discount_Percent) as Avg_Discount_Percentage
             FROM retail 
             Group By Region;''')

df=pd.read_sql_query(query,sql_engine)
df
```

```python
#9. Find the product category with the highest total profit

query = text(''' SELECT DISTINCT
             Sub_Category as Product, max(Profit) as Total_Profit
             FROM retail 
             Group By Product Order By Total_Profit Desc limit 1;''')

df=pd.read_sql_query(query,sql_engine)
df
```

```python
#10. Calculate the total revenue generated per year

query = text(''' SELECT DISTINCT
             Year(Order_Date) as Year , sum(Profit) as Total_Revenue
             FROM retail 
             Group By Year;''')

df=pd.read_sql_query(query,sql_engine)
df
```

```python
#11 List All Orders with Product Information
#Write a query to retrieve a list of all orders (Order_Id, Order_Date, Ship_Mode, Segment, Country)
# along with product details (Product_Id, Category, Sales_Price, Quantity).
# Use an INNER JOIN between the Orders and Product tables.

query = text('''SELECT o.order_Id, o.Order_Date, o.Ship_Mode, o.Segment, o.Country,p.Product_Id, p.Category,p. Sales_Price,p. Quantity
             FROM ORDERS o INNER JOIN PRODUCT p
             ON o.Order_Id = p.Order_Id''')

df= pd.read_sql_query(query,sql_engine)
df
```

```python
#2. Get Total Revenue from Each Product
#Write a query to calculate the total revenue (Sales_Price * Quantity) for each product (Product_Id, Product_Entry_Id).
# Join the Product table with the Orders table to get the total revenue per product.

query = text('''SELECT P. Product_Entry_Id, P.Product_Id, SUM(Sales_Price* Quantity) as total_revenue
             FROM Product p INNER JOIN orders o
             ON o.Order_Id = p.Order_Id
             group by p.Product_Entry_Id, P.Product_Id''')

df= pd.read_sql_query(query,sql_engine)
df
```

```python
# 3. Get All Products Ordered by a Specific Customer
#Write a query to list all products (Product_Id, Category, Sales_Price) ordered by customers in the City 'Henderson'. 
#Join the Orders table with the Product table and filter the City from the Orders table.

query = text('''SELECT  
             P.Product_Id,
             P.Category,
             P.Sales_Price,
             o.city
             FROM Orders o INNER JOIN Product P
             ON o.Order_Id = p.Order_Id
             WHERE o.city = 'Henderson';''')

df= pd.read_sql_query(query,sql_engine)
df.head()
```

```python
#4. Get the Most Expensive Product Ordered
#Write a query to find the most expensive product ordered, showing the Product_Id, Sales_Price, and Order_Date by joining the Orders and Product tables.

query = text('''SELECT  
             P.Product_Id,
             P.Sales_Price,
             o.Order_date
             FROM Orders o INNER JOIN Product P
             ON o.Order_Id = p.Order_Id
             Order by Sales_Price Desc
             Limit 1;''')

df= pd.read_sql_query(query,sql_engine)
df.head()

```

```python
#5. List Orders with Product Details and Total Price
#Write a query to get the order details (Order_Id, Order_Date, Ship_Mode, Sales_Price, Quantity) and 
# calculate the total price per order (total = Sales_Price * Quantity).
# Use an INNER JOIN between the Orders and Product tables.

query = text('''SELECT  
             o.Order_Id,
             o.Order_date,
             o.Ship_Mode,
             p.Sales_Price,
             p.Quantity,
             (p.Sales_Price * Quantity) as total_price
             FROM Orders o INNER JOIN Product P
             ON o.Order_Id = p.Order_Id;''')

df= pd.read_sql_query(query,sql_engine)
df.head()
```

```python
# 6. List All Orders with Product Information (Using LEFT JOIN)
#LEFT JOIN ensures that all records from the Orders table are returned even if there's no matching record in the Product table.

query = text('''SELECT  
            o.Order_Id, 
            o.Order_Date, 
            o.Ship_Mode, 
            o.Segment, 
            o.Country, 
            p.Product_Id, 
            p.Category, 
            p.Sales_Price, 
            p.Quantity
             FROM Orders o LEFT JOIN Product P
             ON o.Order_Id = p.Order_Id;''')

df= pd.read_sql_query(query,sql_engine)
df.head()
```

```python
#7. Get the Most Expensive Product Ordered (Using LEFT JOIN)
#LEFT JOIN to find the most expensive product, even if the product does not have matching orders.

query = text('''SELECT  
            p.Product_Id, 
            p.Sales_Price,
            o.Order_Date 
            FROM Orders o LEFT JOIN Product P
            ON o.Order_Id = p.Order_Id
            Order By p.Sales_Price Desc 
            Limit 1;''')

df= pd.read_sql_query(query,sql_engine)
df
```

```python
# 8. Get Products with Profit Greater than 1000 (Using RIGHT JOIN)
#RIGHT JOIN will ensure that products with profits greater than 10000 are returned, even if there are no matching orders.

query = text('''SELECT  
            p.Product_Id, 
            p.Sales_Price
            FROM Orders o RIGHT JOIN Product P
            ON o.Order_Id = p.Order_Id
            wHERE Sales_Price >= 10000;''')

df= pd.read_sql_query(query,sql_engine)
df
```

```python
#9. Get All Orders from 2023 and 2024
#Use UNION to combine orders from two different years.

query = text('''SELECT  Order_id, Order_Date FROM Orders WHERE YEAR(Order_Date) = 2022
             Union 
             SELECT  Order_id, Order_Date FROM Orders WHERE YEAR(Order_Date) = 2023''')

df= pd.read_sql_query(query,sql_engine)
df
```

```python
# 10. Find All Products in Orders from the ‘South’ Region (Using CROSS JOIN)
#CROSS JOIN creates a Cartesian product between Orders and Product tables and can be used here to find all combinations of orders 
# and products for the 'South' region. 

query = text('''SELECT  
            P.Product_Id,
            P.Category,
            P.Sales_Price
            FROM Orders o CROSS JOIN Product P
            WHERE o.Region  = "South" AND o.Order_Id = p.Order_Id;''')

df= pd.read_sql_query(query,sql_engine)
df
```
